```python
from scipy.stats import binom

# setting the values
# of n and p
n = 6
p = 0.6

# defining the list of r values
r_values = list(range(n + 1))
# obtaining the mean and variance 
mean, var = binom.stats(n, p)
# list of pmf values
dist = [binom.pmf(r, n, p) for r in r_values ]

# printing the table
print("r\tp(r)")
for i in range(n + 1):
    print(str(r_values[i]) + "\t" + str(dist[i]))
# printing mean and variance
print("mean = "+str(mean))
print("variance = "+str(var))

```

```python
#A coin is tossed four times. Calculate the probability of obtaining more heads than tails.


from scipy.stats import binom

# setting the values
# of n and p
n = 4
p = 0.5

# defining the list of r values
r_values = [3,4]
# obtaining the mean and variance 
mean, var = binom.stats(n, p)
# list of pmf values
dist = [binom.pmf(r, n, p) for r in r_values ]
tot=0
# printing the table
print("r\tp(r)")
for i in range(len(r_values)):
    tot+= dist[i]
    print(str(r_values[i]) + "\t" + str(dist[i]))
# printing mean and variance
print("mean = "+str(mean))
print("variance = "+str(var))
print("tot = "+str(tot))
```

```python
#There are 10 red and 20 blue balls in a box. A ball is chosen at random and it is noted whether it is red.
#The process repeats, returning the ball 10 times. Calculate the expected value and the standard deviation of this game.


from scipy.stats import binom
import math
# setting the values
# of n and p
n = 10
p = 1/3

# defining the list of r values
r_values = list(range(n + 1))
# obtaining the mean and variance 
mean, var = binom.stats(n, p)
# list of pmf values
std= math.sqrt(var)

dist = [binom.pmf(r, n, p) for r in r_values ]
tot=0
# printing the table
print("r\tp(r)")
for i in range(len(r_values)):
    print(str(r_values[i]) + "\t" + str(dist[i]))
# printing mean and variance
print("mean = "+str(mean))
print("variance = "+str(var))
print("Standard deviation = "+str(std))


```
