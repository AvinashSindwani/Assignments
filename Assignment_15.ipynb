```python
#Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to its next() method: 2, 3, 5, 7, 11, ...

def prime(num):
    st={2}
    for i in range(2,num):
        cnt=0
        for j in range(2,i):
            if i % j == 0:
                cnt+=1
                break
        if cnt == 0:
            st.add(i)
    return st;

prime(100)


def genPrimes():
    primes = [2]
    yield primes[0]
    guess = 3
    while True:
        if all(guess%x != 0 for x in primes):
            primes.append(guess)        
        if guess == primes[-1]:
            yield primes[-1]
        guess += 2
for i in genPrimes():
    print(i)
    
```
